## AlexNet
- AlexNet赢了2012年ImageNet竞赛
- 更大更深的LeNet
- 主要改进：
    - 丢弃法（Dropout）
    - ReLu （避免梯度消失、更平滑）
    - MapPooling
- 计算机视觉方法论的改变（将计算机视觉中的主要工作从人工特征提取转为通过CNN学习特征）

## AlexNet架构


AlexNet比LeNet深很多，AlexNet输出分类为1000

### 更多细节
- 激活函数从sigmoid变成了ReLu（减缓梯度消失）
- 隐藏全连接层后加入了丢弃层
- 数据增加（随机增加光照、颜色、色温等特征）

AlexNet参数个数是LeNet的11x， 一次前向运算AlexNet的FLop（约1G）是LeNet的250x

## 总结
- AlexNet是更大更深的LeNet，10X参数个数，260x计算复杂度
- 新加入了丢弃发，ReLU，最大池化层和数据增强
- AlexNet赢下了2012 ImageNet竞赛后，标志着新的一轮神经网路热潮的开始
