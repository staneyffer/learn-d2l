# 数据增广
## 数据增强
- 增加一个已有的数据集，是的有更多的多样性
    - 在语言里加入各种不同的背景噪音
    - 改变图片的颜色和形状

## 使用增强数据训练
一般在原始数据上，训练时在线随机生成增强数据

## 常用的增强手段
- 左右翻转
- 上下翻转
    - 不总是可行

## 切割
- 从图片中切割一块，然后变形到固定形状
    - 随机高宽比（e.g. [3/4, 4/3]
    - 随机大小 (e.g. [8%, 100%])
    - 随机位置切割

## 颜色
- 改变色调，饱和度，明亮度（e.g.[0.5, 1.5])


## 总结
- 数据增广通过变形数据来获取多样性从而使得模型泛化性更好
- 常见图片增广包括翻转、切割、变色
