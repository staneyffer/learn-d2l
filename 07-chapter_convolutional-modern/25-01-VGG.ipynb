## 使用块的网络(VGG)

- AlexNet比LeNet更深更大来获得更好的精度
- 能不能更深更大
- 选项
    - 更多的全连接层（太贵）
    - 更多的卷积层
    - 将卷积层组合成块

## VGG块

- 深 VS. 宽？
    - $5 \times 5$卷积
    - $3 \times 5$卷积
    - 深但窄效果更好
- VGG块
    - $3 \times 3$卷积（n层，m通道）
    - $2 \times 2$最大池化层（步幅2）
    



## VGG架构
- 多个VGG块后接全连接
- 不同次数的重复块得到不同的架构(VGG-16, VGG-19)


## 进度

- LeNet（1995）
    - 2卷积+ 池化层
    - 2全连接层
- AlexNet
    - 更大更深
    - ReLU， Dropout，数据增强
- VGG
    - 更大更深的AlexNet（重复的VGG块）

## 总结
-  VGG使用可重复使用的卷积块来构建深度卷积神经网络
- 不同的卷积块个数和超超参数可以得到不同复杂度的变种
