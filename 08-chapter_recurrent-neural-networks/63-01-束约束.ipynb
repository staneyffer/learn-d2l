## 束约束（Beam Search）

## 贪心搜索
- 在seq2seq中我们是用贪心搜索来预测序列
    - 将当前时刻预测概率最大的词输出
    
- 但贪心可能不是最优的



## 穷举搜索

- 最优算法：对所有可能的序列，计算它的概率，然后选取最好的那个
- 如果输出字典大小为$n$，序列最长为$T$，那么我们需要考察$n^T$个序列
    - $n = 10000, T = 10: n^T = 10^{40}$
    - 计算上不可取

## 束约束
- 保存最好的K个候选
- 在每个时刻，对每个候选新加一项（$n$种可能），在$kn$个候选种选出最好的$k$个



## 复杂度

- 时间复杂度$O(knT)$
    - $k = 5, n = 100000, T = 10: KnT = 5 \times 10^5$
- 每个候选的最终分数是（长句子概率更低）：
$$
\frac{1}{L^{\alpha}} \text{log}p(y1, ..., y_L) = \frac{1}{L^{\alpha}} \sum_{t^{'} = 1}^L \text{log} p(y_{t^{'}}|y_1, ...,y_{t^{'} -1})
$$
其中$\frac{1}{L^{\alpha}}$用来给长句子加权， 通常$\alpha = 0.75$

## 总结
- 束约束在每次搜索时保存$k$个最好的候选
    - k = 1时是贪心搜索
    - k一般取5或者10
