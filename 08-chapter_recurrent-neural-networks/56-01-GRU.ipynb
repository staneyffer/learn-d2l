## 候选隐状态


这里的的$\pmb R_t \odot \pmb H_{t-1}$ 是充值门与上一刻的隐状态的哈达玛积（按元素乘法），其中$\pmb R_t$是$0-1$之间的数，当$\pmb R_t$接近0时，表示尽可能忽视上一刻的隐藏状态。

## 隐状态



其中，$\pmb Z_t$是更新门（update gate），如果$\pmb Z_t = 1$则表示直接使用上一刻的隐状态，即忽视当前的。如果$\pmb Z_t = 0$表示基本不看过去的状态，只看现在的状态，回到的RNN的情况。
