## 单隐藏层

在输入层和输出层之间加入一个隐藏层

## 单隐层藏-单分类

- ## 输入 $\pmb x \in \mathbb R^n$
- ## 隐藏层 $\pmb W_1 \in \mathbb R^{m \times n}$, $\pmb b_1 \in \mathbb R^m$
- ## 输出层 $w_2 \in \mathbb R^m, b_2 \in \mathbb R$
$$
\pmb h = \sigma(\pmb W_1 \pmb x + \pmb b_1) \\
o = \pmb w_2^T \pmb h + b_2 \tag{1.1}
$$ 
$\sigma$是按元素的激活函数

为什么非线性的激活函数?

假设激活函数$\sigma(x) = x$， 代入上式1.1，得

$$
\pmb h = \pmb W_1 \pmb x + \pmb b_1 \\
o = \pmb w_2 ^T \pmb h + b_2 \\
o = \pmb w_2^T \pmb W_1 \pmb x + b{'}
$$
任然是线性模型(因为: $\pmb w_2^T$是一个向量)

## Sigmoid激活函数

将输入投影到$(0, 1)$
$$
\text{sigmoid}(x) = \frac{1}{1 + \text{exp}(-x)}
$$

```python
import torch
from d2l import torch as d2l
import numpy as np

x = np.arange(-10, 10,1)

def f(x):
    return 1 / (1 + np.exp(-x))

d2l.plot(x, [f(x)],'x', 'f(x)')
```

## Tanh激活函数

将输入投影到(-1, 1)
$$
\text{tanh}(x) = \frac{1 - \text{exp}(-2x)}{1 + \text{exp}(-2x)}
$$

```python
def f2(x):
    return (1 - np.exp(-2 * x)) / (1 + np.exp(-2*x))

d2l.plot(x, [f2(x)], 'x', 'tanh(x)')
```

## ReLU激活函数（计算简单）

ReLU: rectified linear unit

$$
\pmb {\text{ReLU}}(x) = \text{max}(x, 0)
$$

```python
def ReLU(x):
    return x * (x > 0)
d2l.plot(x, [ReLU(x)], 'x', 'ReLu(x)')
```

## 多类分类
- ### 输入 $\pmb x \in \mathbb R^n$
- ### 隐藏层 $\pmb W_1 \in \mathbb R^{m \times n}, \pmb b_1 \in \mathbb R^m$
- ### 输出层 $\pmb W_2 \in \mathbb R^{m \times k}, \pmb b_2 \in \mathbb R^k$
$$
\pmb h = \sigma(\pmb W_1 x + \pmb b_1) \\
\pmb o = \pmb W_2^T h + \pmb b_2 \\
\pmb y = \text{softmax}(\pmb o)
$$

## 多隐藏层
$$
\pmb h_1 = \sigma(\pmb W_1\pmb x + \pmb b_1) \\
\pmb h_2 = \sigma(\pmb W_2\pmb h_1 + \pmb b_2) \\
\pmb h_3 = \sigma(\pmb W_3\pmb h_2 + \pmb b_3) \\
\pmb o = \pmb W_4 \pmb h_3 + \pmb b_4
$$

### 超参数
- ### 隐藏层参数
- ### 每层隐藏层的大小（一般情况，离输出层越近，层数越小）

##  总结
- ### 多层感知机使用隐藏层和激活函数来得到非线性模型
- ### 常用激活函数是Sigmoid, Tanh, ReLU（没什么特别的选择就选ReLU）
- ### 使用Softmax来处理多类分类
- ### 超参数为隐藏层数，和各层大小
